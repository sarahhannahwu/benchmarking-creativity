```python
import json

# Filter out errors, keep only json strings
df['data_type'] = df['response'].apply(lambda x: str(type(x)))
not_str_mask = df['data_type'].str.contains("str")
subset = df[not_str_mask].copy()

# Convert json strings to dictionaries
subset['response_dict'] = subset['response'].apply(json.loads) 
display(subset.head())
display(subset['response_dict'].iloc[0])
responses = pd.json_normalize(subset['response_dict'], record_path=['items'])
responses
```

```python

df_list = []
for idx, row in subset.iterrows():
    _response = pd.json_normalize(row['response_dict'], record_path=['items'])
    _response['timestamp'] = row['timestamp']
    _response['prompt_file'] = row['prompt_file']
    _response['model'] = row['model']
    df_list.append(_response)
responses = pd.concat(df_list)
responses
```
